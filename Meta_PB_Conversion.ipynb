```python
import tensorflow as tf
tf.disable_v2_behavior()
tf.compat.v1.disable_eager_execution()

meta_path = '../Personal/cat-vs-dog/trained_models/Model-DropoutLayer.ckpt' # Your .meta file
output_node_names = ['output_layer']    # Output nodes

with tf.Session() as sess:
    saver = tf.train.import_meta_graph(meta_path)
    saver.restore(sess,tf.train.latest_checkpoint('../Personal/cat-vs-dog/trained_models/'))
    frozen_graph_def = tf.graph_util.convert_variables_to_constants(
        sess,
        sess.graph_def,
        output_node_names)
    with open('Ajith_output_graph.pb', 'wb') as f:
        f.write(frozen_graph_def.SerializeToString())
```
